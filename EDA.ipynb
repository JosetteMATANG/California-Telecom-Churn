## Variables quantitatives

```python
# Sélection des colonnes quantitatives
quanti = base.select_dtypes(include=["int64", "float64"]).columns
# Exclusion de Latitude, Longitude, et Zip Code
quanti = quanti.drop(['Latitude', 'Longitude', 'Zip Code'])

# Tracer un boxplot pour chaque variable
n_cols = 4  # nombre de boxplots par ligne
n_rows = (len(quanti) + n_cols - 1) // n_cols  # Calcul automatique du nombre de lignes

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 5*n_rows))
axes = axes.flatten()  # Pour accéder facilement aux axes

for i, col in enumerate(quanti):
    sns.boxplot(y=base[col], ax=axes[i])
    axes[i].set_title(col)
    axes[i].grid(True)

# Supprimer les sous-graphiques vides (s'il y en a)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.show()

```

```python
corr_matrix = base.loc[:, quanti].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".3f", cmap="Blues")

plt.title('Matrice de Corrélation', fontsize=16)
plt.show()
```

## Variable cible

```python
#donut de Customer Status

# Compter les catégories de Customer Status
customer_status_counts = base['Customer Status'].value_counts()

# Création du camembert
fig, ax = plt.subplots(figsize=(8, 8))

# Tracer un pie chart
wedges, texts, autotexts = ax.pie(
    customer_status_counts,
    labels=customer_status_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width':0.4},
    colors=plt.cm.Set3.colors
)

# Cercle au centre pour accentuer l'effet donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Titre
ax.set_title('Répartition de Customer Status', fontsize=16)

plt.show()
```
